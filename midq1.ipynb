```python
import matplotlib.pyplot as plt
import json

dic = {}

f = open("numbers.txt", "r")
for line in f:
    line = line.strip()
    if line:
        words = line.split(",")
        for word in words:
            num = int(word)
            if num not in dic:
                dic[num] = 0

            dic[num] += 1
f.close()


for num in sorted(dic):
    freq = dic[num]
    print(f"{num} : {freq}")



x = list(range(10))
freq = [dic[i] for i in x]
plt.bar(x=x, height=freq )
plt.grid(True)
plt.xlabel("index")
plt.ylabel("frequency")
plt.title("frequency of each number")
plt.show()



json_str = json.dumps(dic)
f = open("freq.json", "w")
f.write(json_str)
f.close()
```
